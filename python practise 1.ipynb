3.Check whether a number is prime



```python
n = int(input("Enter the number: "))

if n <= 1:
    print("Not a prime number")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

5.Reverse a number





```python
n = int(input("Enter number: "))

rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reversed number:", rev)
```

```python
n = int(input("Enter number: "))

even_count = 0
odd_count = 0

while n > 0:
    digit = n % 10

    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

    n = n // 10

print("Even digits:", even_count)
print("Odd digits:", odd_count)
```

```python
n = int(input())
row = [1]

for i in range(n):
    print(row)
    new_row = [1]
    for j in range(len(row)-1):
        new_row.append(row[j] + row[j+1])
    new_row.append(1)
    row = new_row
```

```python
n = int(input())

for i in range(1, n+1):
    for j in range(1, 2*n):
        if j == n-i+1 or j == n+i-1 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

45.Give code for prime number if it is prime give its count and if not prime simply print it is not prime

```python
n = int(input("enter the number: "))


if n <= 1:
    print("Not a prime number")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if not is_prime:
        print("Not a prime number")
    else:
        
        prime_count = 0

        for num in range(2, n + 1):
            divisor_count = 0

            for j in range(1, num + 1):
                if num % j == 0:
                    divisor_count += 1

            if divisor_count == 2:
                prime_count += 1

        print(n, "is a prime number and it is the", prime_count, "prime number")
```
