```python
#Factorial of a number
n = int(input("ENter a number"))
fact = 1
if n == 0 or n == 1:
  print("Fact:",fact)
for i in range(1,n+1):
  fact*=i
print(fact)
```

```python
# Reverse of number
n =int(input("Enter a number:"))
rev = 0
while n>0:
  digit = n %10
  rev = rev*10 + digit
  n = n//10
print(rev)
```

```python
#Fibonacci series (N terms)
n = int(input("Enter a number"))
a,b = 0,1
for _ in range(n):
  print(a,end="")
  a,b = b,a+b
```

```python
#Check palindrome number
n = int(input("Enter a number "))
rev = 0
temp = n
while n>0:
  rev = rev*10 + n%10
  n //=10
if temp == rev:
  print(" it is palindrome")
else:
  print("Not palindrome")
```

```python
# GCD of number
a,b = map(int,input("Enter two number").split())
while b!=0:
  a,b = b,a%b
print("GCD:",a)
```

```python
# LCM
a, b = map(int, input("Enter two numbers: ").split())
x, y = a, b

while y != 0:
    x, y = y, x % y

gcd = x
lcm = (a * b) // gcd

print("LCM:", lcm)
```

```python
#Check palindrome string
s = input("enter a string")
rev = ""
for i in s:
  rev=i+rev
if s == rev:
  print("it is palindrome")
else:
  print("it is not palindrom")
```

```python
#Count Vowels and Consonants
s = input("Enter string: ").lower()
vowels = 0
consonants = 0

for ch in s:
    if ch.isalpha():
        if ch in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
#Check Anagram
s1 = input("Enter a string: ").replace(" ","").lower()
s2 = input("ENter a string: ").replace(" ","").lower()
if sorted(s1) == sorted(s2):
  print("Anagram")
else:
  print("Not Anagram")
```

```python
# words in a string
s = input("Enter string: ").strip()
count = 0
in_word = False

for ch in s:
    if ch != " " and not in_word:
        count += 1
        in_word = True
    elif ch == " ":
        in_word = False

print("Word count:", count)
```
